# Reading in updated dataframe

```python
all_data = pd.read_csv('all_data.csv')
```

### Clean up the data

##### drop rows of NaN

```python
nan_df =  all_data[all_data.isna().any(axis=1)]
```

```python
all_data = all_data.dropna(how='all')
all_data.head()
```

```python
nan_df =  all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

### Find 'Or' and Delete it

```python
import re
all_data[all_data['Order Date'].str.contains('Or')]
```

```python
# or second way
temp_df = all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df.head()
```

```python
#delete rows that contains "OR"
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

### Task 3:Add sales column

```python
#Before we need to convert all columns in correct types
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Add month column

```python
all_data['Month'] =all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

# Question 1:What was the best month for sales? How much that month was earned?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month')
plt.show()
```

# Question 2: What city had the highest number of sales?

#### Task 4:Add a column a city

```python
all_data.head()
```

```python
# let's use apply()
def get_city(address):
    return address.split(',')[1]

def get_state(s):
    return s.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + '(' + get_state(x) + ')')


all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
cities=[city for city,df in all_data.groupby('City')]
plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical')
plt.xlabel('Cities')
plt.ylabel('Sales in USD($)')
plt.show()
```

### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour,df in all_data.groupby('Hour')]

plt.plot(hours,all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```
